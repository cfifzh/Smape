### –ü—Ä–æ–≤–µ—Ä–∫–∞ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –æ–±–æ–≥–∞—â–µ–Ω–∏—è

–≥–¥–µ:
* C ‚Äî –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –≤ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–µ –ø–æ—Å–ª–µ —Ñ–ª–æ—Ç–∞—Ü–∏–∏/–æ—á–∏—Å—Ç–∫–∏;
* F ‚Äî –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –≤ —Å—ã—Ä—å–µ/–∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–µ –¥–æ —Ñ–ª–æ—Ç–∞—Ü–∏–∏/–æ—á–∏—Å—Ç–∫–∏;
* T ‚Äî –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –≤ –æ—Ç–≤–∞–ª—å–Ω—ã—Ö —Ö–≤–æ—Å—Ç–∞—Ö –ø–æ—Å–ª–µ —Ñ–ª–æ—Ç–∞—Ü–∏–∏/–æ—á–∏—Å—Ç–∫–∏.

$\displaystyle Recovery = \frac{C \cdot (F-T)} {F \cdot (C-T)} \cdot100$

<div class="alert alert-success">
<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ v2<a class="tocSkip"> </h2>
<hr>
<b>–í–µ—Ä–Ω–æ!üëç:</b> –°–æ–≥–ª–∞—Å–µ–Ω, –∏–∑–æ–±—Ä–µ—Ç–∞—Ç—å –≤–µ–ª–æ—Å–∏–ø–µ–¥ –Ω–µ —Å—Ç–æ–∏—Ç, –Ω–æ –∫–∞—Ä—Ç–∏–Ω–∫–∞ –±—É–¥–µ—Ç –Ω–µ –≤—Å–µ–≥–¥–∞, –ø–æ—ç—Ç–æ–º—É –Ω–∞ –≤—Å—è–∫–∏–π —Å–ª—É—á–∞–π –æ—Å—Ç–∞–≤–∏–ª —Å—Å—ã–ª–∫—É (–∫–æ—Ç–æ—Ä–∞—è –º–æ–∂–µ—Ç –ø—Ä–∏–≥–æ–¥–∏—Ç—å—Å—è —É–∂–µ –Ω–∞ —Å–ª–µ–¥—É—é—â–µ–º —Å–ø—Ä–∏–Ω—Ç–µ) :)
</div>

```python
recovery_tab = train[['rougher.output.concentrate_au', 'rougher.input.feed_au', 
                      'rougher.output.tail_au', 'rougher.output.recovery']].copy()

C = recovery_tab['rougher.output.concentrate_au']
F = recovery_tab['rougher.input.feed_au']
T = recovery_tab['rougher.output.tail_au']

recovery_tab['rougher.output.recovery_formula'] = C * (F - T) / F / (C - T) * 100

# –†–∞—Å—á–∏—Ç–∞–µ–º MAE
mean_absolute_error(recovery_tab['rougher.output.recovery'], recovery_tab['rougher.output.recovery_formula'])
```

C—Ä–µ–¥–Ω—è—è –∞–±—Å–æ–ª—é—Ç–Ω–∞—è –æ—à–∏–±–∫–∞ —Å—Ç—Ä–µ–º–∏—Ç—å—Å—è –∫ –Ω—É–ª—é. –î–∞–Ω–Ω—ã–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã–µ –≤ —Å—Ç–æ–ª–±—Ü–µ rougher.output.recovery –≤–∞–ª–∏–¥–Ω—ã.
–ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –Ω–∏—Ö –±–æ–ª–µ–µ –≤–Ω–∏–º–∞—Ç–µ–ª—å–Ω–æ. –ü–æ–ø—Ä–æ–±—É–µ–º –≤—ã—è–≤–∏—Ç—å –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ –∑–∞–∫–æ–Ω–æ–º–µ—Ä–Ω–æ—Å—Ç–∏.

<div class="alert alert-success">
<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ <a class="tocSkip"> </h2>
<hr>
<b>–í–µ—Ä–Ω–æ!üëç:</b> –§–æ—Ä–º—É–ª–∞ –≤ —è—á–µ–π–∫–µ —Å –∫–æ–¥–æ–º –Ω–∞–≥–ª—è–¥–Ω–æ –∑–∞–ø–∏—Å–∞–Ω–∞, –≤—ã–≤–æ–¥ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π, –º–æ–ª–æ–¥–µ—Ü.
</div>

```python
null_index = []
match_count = 0
match_acceptable_count =  0
match_no_index = []
for i in range(len(recovery_tab)):
    
    if (recovery_tab['rougher.output.recovery'].loc[i] == 0 and
        recovery_tab['rougher.output.recovery_formula'].loc[i] == 0):
        null_index.append(i)
    
    elif recovery_tab['rougher.output.recovery'].loc[i] == recovery_tab['rougher.output.recovery_formula'].loc[i]:
        match_count += 1
        
    elif (abs(1 - 
         (recovery_tab['rougher.output.recovery'].loc[i] / recovery_tab['rougher.output.recovery_formula'].loc[i])
        ) < 0.000001):
        match_acceptable_count += 1
    
    else:
        match_no_index.append(i)
        
(
    print(f'–ó–Ω–∞—á–µ–Ω–∏–µ —è—á–µ–π–∫–∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —Ñ–æ—Ä–º—É–ª–µ –∏ —Ä–∞–≤–Ω–æ 0 –≤ \033[1m{len(null_index)}\033[0m —Å—Ç—Ä–æ–∫–∞—Ö;\n' 
          f'–ó–Ω–∞—á–µ–Ω–∏–µ —è—á–µ–π–∫–∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —Ñ–æ—Ä–º—É–ª–µ –∏ –Ω–µ —Ä–∞–≤–Ω–æ 0 –≤ \033[1m{match_count}\033[0m —Å—Ç—Ä–æ–∫–∞—Ö;\n'
          f'–ó–Ω–∞—á–µ–Ω–∏–µ —è—á–µ–π–∫–∏ –æ—Ç–ª–∏—á–∞—Ç—Å—è –æ—Ç —Ñ–æ—Ä–º—É–ª—ã –º–µ–Ω–µ —á–µ–º –Ω–∞ 0.000001 –≤ \033[1m{match_acceptable_count}\033[0m —Å—Ç—Ä–æ–∫–∞—Ö;\n'
          f'–ó–Ω–∞—á–µ–Ω–∏–µ —è—á–µ–π–∫–∏ –æ—Ç–ª–∏—á–∞—Ç—Å—è –æ—Ç —Ñ–æ—Ä–º—É–ª—ã –±–æ–ª–µ–µ —á–µ–º –Ω–∞ 0.000001 –≤ \033[1m{len(match_no_index)}\033[0m —Å—Ç—Ä–æ–∫–∞—Ö;\n')
)

if (len(null_index) + match_count + match_acceptable_count + len(match_no_index)) == train.shape[0]:
    print('\033[1m–ó–Ω–∞—á–µ–Ω–∏–µ —è—á–µ–µ–∫ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —Ñ–æ—Ä–º—É–ª–µ\033[0m')
else:
    print( f'–ü—Ä–æ–≤–µ—Ä—è–µ–º —Å–ª–µ–¥—É—é—â–∏–µ —Å—Ç—Ä–æ–∫–∏ {match_no_index}\n')
```

### C—É–º–º–∞—Ä–Ω–∞—è –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏—è –≤—Å–µ—Ö –≤–µ—â–µ—Å—Ç–≤ –Ω–∞ —Ä–∞–∑–Ω—ã—Ö —Å—Ç–∞–¥–∏—è—Ö

–í—ã–¥–µ–ª–∏–º –Ω—É–∂–Ω—ã–µ —Å—Ç–æ–ª–±—Ü—ã

```python
metal_list_2 = ['au', 'ag', 'pb', 'sol']                #–Ω–∞—à–∏ –≤–µ—â–µ—Å—Ç–≤–∞
metal_list_remove = ['tail', 'calculation', '_pb_']     #–ª–∏—à–Ω–∏–µ –∫–æ–ª–æ–Ω–∫–∏
concentrate_col = []                                    #—Å–ø–∏—Å–æ–∫ –¥–ª—è –Ω—É–∂–Ω—ã—Ö –∫–æ–ª–æ–Ω–æ–∫

for col in full.columns.to_list():                      #–ø–µ—Ä–µ–±–æ—Ä —Å –ø–æ–∏—Å–∫–æ–º —É–ø–æ–º–∏–Ω–∞–Ω–∏—è –≤–µ—â–µ—Å—Ç–≤ –≤ –Ω–∞–∑–≤–∞–Ω–∏—è—Ö –∫–æ–ª–æ–Ω–æ–∫ 
    for metal in metal_list_2:
        if metal in col:
            concentrate_col.append(col)
            
    for metal in metal_list_remove:                     #—É–¥–∞–ª–µ–Ω–∏–µ –ª–∏—à–Ω–∏—Ö –∫–æ–ª–æ–Ω–æ–∫
        if metal in col:
            concentrate_col.remove(col)
               
concentrate_col                                         #–ø—Ä–æ–≤–µ—Ä–∫–∞
```

### –§—É–Ω–∫—Ü–∏—é –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –∏—Ç–æ–≥–æ–≤–æ–π sMAPE

* **≈∑<sub>i</sub>** - –∑–Ω–∞—á–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –¥–ª—è –æ–±—ä–µ–∫—Ç–∞ —Å –ø–æ—Ä—è–¥–∫–æ–≤—ã–º –Ω–æ–º–µ—Ä–æ–º i

<div class="alert alert-info">
    
<b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞:</b>
    
–°–µ–π—á–∞—Å –ø—Ä–∏—Å—Ç—É–ø–ª—é –∫ –æ–±—É—á–µ–Ω–∏—é –º–æ–¥–µ–ª–µ–π –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –≤ –æ—Ç–¥–µ–ª—å–Ω–æ—Å—Ç–∏, —Å—Ç–∞—Ä—ã–π –∫–æ–¥ —É–¥–∞–ª—è—Ç—å –Ω–µ –±—É–¥—É, —á—Ç–æ–±—ã –º–æ–∂–Ω–æ –±—ã–ª–æ —Å—Ä–∞–≤–Ω–∏—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç. –ò–∑–≤–∏–Ω–∏, —á—Ç–æ –±—É–¥—É –¥–µ–ª–∞—Ç—å –±–µ–∑ –∞–≤—Ç–æ–º–∞—Ç–∏–∑–∞—Ü–∏–∏.</div>

<div class="alert alert-success">
<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ v2<a class="tocSkip"> </h2>
<hr>
<b>–í–µ—Ä–Ω–æ!üëç:</b> –í–∏–∂—É, —á—Ç–æ —Ç—ã —Å –∞–≤—Ç–æ–º–∞—Ç–∏–∑–∞—Ü–∏–µ–π —Ö–æ—Ä–æ—à–æ —Å–ø—Ä–∞–≤–ª—è–µ—à—å—Å—è, –º–æ–∂–µ—Ç –ø–æ—Ç–æ–º –¥–ª—è —Å–µ–±—è –∫–∞–∫-–Ω–∏–±—É–¥—å –ø—Ä–∏–∫—Ä—É—Ç–∏—à—å :)
</div>

```python
def smape(y_true, y_pred):
    error = abs(y_true - y_pred)
    scale = (abs(y_true) + abs(y_pred)) / 2
    
    result = (error / scale * 100).mean()
    return result

def final_smape(sample_true, sample_pred):
    rougher = smape(sample_true.iloc[:, 0], sample_pred[:,0])
    final = smape(sample_true.iloc[:, 1], sample_pred[:,1])
    return 0.25 * rougher + 0.75 * final
```
