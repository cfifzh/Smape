### Проверка эффективности обогащения

Проверим эффективность обогащения rougher.output.recovery и рассчитанную по формуле:
![title](https://pictures.s3.yandex.net/resources/Recovery_1576238822.jpg)
где:
* C — доля золота в концентрате после флотации/очистки;
* F — доля золота в сырье/концентрате до флотации/очистки;
* T — доля золота в отвальных хвостах после флотации/очистки.

$\displaystyle Recovery = \frac{C \cdot (F-T)} {F \cdot (C-T)} \cdot100$

```python
recovery_tab = train[['rougher.output.concentrate_au', 'rougher.input.feed_au', 
                      'rougher.output.tail_au', 'rougher.output.recovery']].copy()

C = recovery_tab['rougher.output.concentrate_au']
F = recovery_tab['rougher.input.feed_au']
T = recovery_tab['rougher.output.tail_au']

recovery_tab['rougher.output.recovery_formula'] = C * (F - T) / F / (C - T) * 100

# Расчитаем MAE
mean_absolute_error(recovery_tab['rougher.output.recovery'], recovery_tab['rougher.output.recovery_formula'])
```

Cредняя абсолютная ошибка стремиться к нулю. Данные представленные в столбце rougher.output.recovery валидны.
Посмотрим на них более внимательно. Попробуем выявить некоторые закономерности.

```python
null_index = []
match_count = 0
match_acceptable_count =  0
match_no_index = []
for i in range(len(recovery_tab)):
    
    if (recovery_tab['rougher.output.recovery'].loc[i] == 0 and
        recovery_tab['rougher.output.recovery_formula'].loc[i] == 0):
        null_index.append(i)
    
    elif recovery_tab['rougher.output.recovery'].loc[i] == recovery_tab['rougher.output.recovery_formula'].loc[i]:
        match_count += 1
        
    elif (abs(1 - 
         (recovery_tab['rougher.output.recovery'].loc[i] / recovery_tab['rougher.output.recovery_formula'].loc[i])
        ) < 0.000001):
        match_acceptable_count += 1
    
    else:
        match_no_index.append(i)
        
(
    print(f'Значение ячейки соответствует формуле и равно 0 в \033[1m{len(null_index)}\033[0m строках;\n' 
          f'Значение ячейки соответствует формуле и не равно 0 в \033[1m{match_count}\033[0m строках;\n'
          f'Значение ячейки отличатся от формулы мене чем на 0.000001 в \033[1m{match_acceptable_count}\033[0m строках;\n'
          f'Значение ячейки отличатся от формулы более чем на 0.000001 в \033[1m{len(match_no_index)}\033[0m строках;\n')
)

if (len(null_index) + match_count + match_acceptable_count + len(match_no_index)) == train.shape[0]:
    print('\033[1mЗначение ячеек соответствует формуле\033[0m')
else:
    print( f'Проверяем следующие строки {match_no_index}\n')
```

Вся разница между данными в ячейках столбцов исходного датафрейма и рассчитанного по формуле связана с округлением, кроме того, обнаружилось некоторое количество нулей. Выясним откуда они появились.

## Модель

### Функцию для вычисления итоговой sMAPE

![title](https://pictures.s3.yandex.net/resources/smape_1576238825.jpg)

где 

* **y<sub>i</sub>** - значение целевого признака для объекта с порядковым номером i в выборке, на которой измеряется качество;
* **ŷ<sub>i</sub>** - значение предсказания для объекта с порядковым номером i

![title](https://pictures.s3.yandex.net/resources/_smape_1576238814.jpg)

```python
def smape(y_true, y_pred):
    error = abs(y_true - y_pred)
    scale = (abs(y_true) + abs(y_pred)) / 2
    
    result = (error / scale * 100).mean()
    return result
```

```python
def final_smape_2(rougher, final):
    return 0.25 * rougher + 0.75 * final
```
